```python
OH_cols = ['street_id', 'metro_dist', 'build_tech', 'area', 
                    'rooms']
#non_categorial_cols = ['date', 'floor', 'n_photos', 'kw1', 'kw2', 'kw3',
#                      'kw4', 'kw5', 'kw6', 'kw7', 'kw8', 'kw9', 
#                       'kw10', 'kw11', 'kw12', 'kw13']

train2_df = train_df.copy()
test2_df = test_df.copy()

train2_df.shape
```

```python
street_mean = train2_df.groupby('street_id').price.mean()
street_mean.plot(figsize = (16, 6))
#street_mean.plot.scatter(y = 'price', figsize = (14, 5))
```

```python
#train2_df = train2_df.drop(['price'], axis = 1)

full_df = pd.concat([train2_df, test2_df], axis = 0)
dfs = []

for col in OH_cols:
    print(f"Encoding {col}")
    df = pd.get_dummies(full_df[col], prefix=col)
    dfs.append(df)

full_df = full_df.drop(OH_cols, axis = 1)
full_df = pd.concat([full_df] + dfs, axis = 1)

encoder = LabelEncoder()

full_df['date'] = encoder.fit_transform(full_df['date'])

train2_df = full_df[full_df.price.notnull()]

test2_df = full_df[full_df.price.isnull()]
test2_df.drop("price", axis = 1, inplace = True)

#train2_df = train2_df.drop(['price'], axis = 1)

train2_df.shape
```

```python
target_col = train2_df.price
train2_df.drop(['price'], axis = 1, inplace=True)
```

```python
train2_df = (train2_df - train2_df.mean(axis = 0)) \
    / train2_df.std(axis = 0)
    
test2_df = (test2_df - test2_df.mean(axis = 0)) \
    / test2_df.std(axis = 0)
```

```python
model = xgb.XGBRegressor(colsample_bytree=0.4, \
                 gamma=0,                 \
                 learning_rate=0.07, \
                 max_depth=3, \
                 min_child_weight=1.5, \
                 n_estimators=10000,    \                                                                
                 reg_alpha=0.75, \
                 reg_lambda=0.45, \
                 subsample=0.6, \
                 seed=42) 
model.fit(train2_df, target_col)
model.boost
```
