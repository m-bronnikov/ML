```python
# RETURN HERE
train_df = train_df.drop(['g_lift', 'street_id'], axis = 1)
test_df = test_df.drop(['g_lift', 'street_id'], axis = 1)
```

```python
OH_cols = ['metro_dist', 'build_tech', 'area', 
                    'rooms']
#non_categorial_cols = ['date', 'floor', 'n_photos', 'kw1', 'kw2', 'kw3',
#                      'kw4', 'kw5', 'kw6', 'kw7', 'kw8', 'kw9', 
#                       'kw10', 'kw11', 'kw12', 'kw13']

train2_df = train_df.copy()
test2_df = test_df.copy()

train2_df.shape
```

```python
full_df = pd.concat([train2_df, test2_df], axis = 0)
dfs = []

for col in OH_cols:
    print(f"Encoding {col}")
    df = pd.get_dummies(full_df[col], prefix=col)
    dfs.append(df)

full_df = full_df.drop(OH_cols, axis = 1)
full_df = pd.concat([full_df] + dfs, axis = 1)

encoder = LabelEncoder()

full_df['date'] = encoder.fit_transform(full_df['date'])

train2_df = full_df[full_df.price.notnull()]

test2_df = full_df[full_df.price.isnull()]
test2_df.drop("price", axis = 1, inplace = True)
```

```python
target_col = train2_df.price
train2_df.drop(['price'], axis = 1, inplace=True)
```

```python
# NORMALISATION

train2_df = (train2_df - train2_df.mean(axis = 0)) \
    / train2_df.std(axis = 0)
    
test2_df = (test2_df - test2_df.mean(axis = 0)) \
    / test2_df.std(axis = 0)
```

```python
# model = xgb.XGBRegressor(colsample_bytree=0.4, \
#                  gamma=0,   \
#                  learning_rate=0.07, \
#                  max_depth=6, \
#                  min_child_weight=1.5, \
#                  n_estimators=1000,                                                                
#                  reg_alpha=0.75, \
#                  reg_lambda=0.45, \
#                  subsample=0.6, \
#                  seed=42) 
# model.fit(train2_df, target_col)

# xgb_clf = xgb.XGBRegressor() # method

steps = 5

first_end = train2_df.shape[0] // 2

dist = train2_df.shape[0] // (2 * steps)

d = {}



parameters = {'n_estimators': [500, 900, 1200, 1600], 'max_depth':[3, 5, 7, 9] }
for i in tqdm(range(steps)):
    X = train2_df.iloc[:first_end + dist * i]
    Y = target_col[:first_end + dist*i]
    X_t = train2_df.iloc[first_end + dist * i : first_end + dist * (i+1)]
    Y_t = target_col[first_end + dist * i : first_end + dist * (i+1)]
    best = 1000000000000000000
    
    for j in parameters['n_estimators']:
        for k in parameters['max_depth']:
            print("step =", i, "n_estimtor =", j, "max_depth =", k)
            xgb_rgr = xgb.XGBRegressor(n_estimators=j, max_depth=k)
            xgb_rgr.fit(X, Y)
            Y_pred = xgb_rgr.predict(X_t)
            num = sm.mean_squared_error(Y_t, Y_pred)
            if num < best:
                best = num
                d[i] = (j, k)
                

    
d
```
