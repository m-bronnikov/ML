##### Задание: найти индексы чисел меньше 4 в массиве и вывести их сумму
`np.where` позволяет получить индексы истинных значений в передаваемом массиве в виде кортежа одномерных массивов. Мы передаем матрицу, значения которой истинны, если в данной позиции значение $<4$. `np.sum` позволяет получить сумму всех значений массива. Тогда ответом является сумма сумм индексов по обоим измерением.

```python
indexies = np.where(A < 4)
summa = np.sum(indexies[0]) + np.sum(indexies[1])
print("Сумма сумм индексов" \
      " между собой:", summa)
```

##### Задание: найти индексы чисел меньше 4 и не меньше 2 в массиве и вывести их сумму
Функция `np.logical_and` позволяет сделать логическое "И" между 2-мя матрицами булевых значений и получить результирующую 3-юю матрицу-маску для нахождения индексов.

```python
mask = np.logical_and(A >= 2, A < 4)
indexies = np.where(mask)
summa = np.sum(indexies[0]) + np.sum(indexies[1])
print("Сумма сумм индексов" \
      " между собой:", summa)
```

##### Задание: найти сумму значений по индексам предыдущего задания
Обращение к элементам массива по маске дает список соответствующих элементов сумму которых нужно найти

```python
summa = np.sum(A[mask])
print("Сумма значений по индексам " \
      "предыдущего задания:", summa)
```
